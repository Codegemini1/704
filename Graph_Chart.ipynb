```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1,2,3,4]
y = [10,20,25,30]
```

```python
plt.plot(x,y)
plt.title('Graph of X and Y')
plt.xlabel('X_AXIS')
plt.ylabel('Y_AXIS')
plt.grid(True)
plt.show()
```

# Bar Graph

```python
x = ['A', 'B', 'C']
y = [10,20,15]
```

```python
plt.bar(x,y,color='skyblue')
plt.title('Bar Plot')
plt.show()
```

```python
plt.bar(x,y,color='m')
plt.title('Bar Plot')
plt.show()
```

# Histogram

```python
import numpy as np
```

```python
df = np.random.randint(0,50,500)
```

```python
plt.hist(df, bins=30, color='purple')
plt.title('Histogram')
plt.show()
```

# Scatter Plot

```python
x = [50,70,80,70]
y = [99,34,54,78]
```

```python
plt.scatter(x,y,color='red')
plt.title('Scatter Plot')
plt.show()
```
